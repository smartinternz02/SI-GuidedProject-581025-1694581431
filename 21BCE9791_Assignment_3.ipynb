Detecting and removing the outliers

```python
sns.boxplot(ds['Fare'],data=ds);
```

```python
q1= ds.Fare.quantile(0.25)
q3= ds.Fare.quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit =q3+1.5*IQR
upper_limit
```

```python
ds = ds[ds.Fare<upper_limit]
```

```python
sns.boxplot(ds['Fare'],data=ds);
```
