```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
df=sns.load_dataset('car_crashes')
df
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df.describe()
```

CORRELATION

```python
corr=df.corr()
print(corr)
```

HEAT MAP

```python
plt.subplots(figsize=(15,15))
sns.heatmap(corr,annot=True);
```

SCATTER PLOT

```python
sns.scatterplot(data=df,x='alcohol',y='speeding',s=100)
plt.xlabel("Alcohol")
plt.ylabel("Speeding")
plt.title("Alcohol vs Speeding");
```

INFERENCE: It can be observed that the more the alcohol more the speeding is.

DISTPLOT

```python
sns.set_style("darkgrid")
sns.distplot(df['total']);
```

```python
df['total'].mean()
```

```python
df['total'].median()
```

INFERENCE: The distribution is skewed 

BARPLOT

```python
plt.figure(figsize=(12,12))
sns.barplot(data=df,x='total',y='abbrev');
```

INFFERENCE: It is observed that the total number of crashes is more at WV,SC and ND

LINE PLOT:

```python
sns.lineplot(data=df,x="alcohol",y="total");
```

INFERENCE: It can be observed that as the alcohol increases the total crashes increases

```python
sns.displot(df['no_previous'],bins=np.arange(6,22,2));
```

INFERENCE: It is observed that between 12 to 14 percent of no_previous have been into most car crashes

REL PLOT

```python
sns.relplot(data=df,x="ins_premium",y="ins_losses",hue="abbrev");
```

INFERENCE: Shows the relation between ins_losses and ins_premium across the countries

PAIR PLOT

```python
sns.pairplot(data=df);
```

INFERENCE: Shows the relation between each variable in the car_crashes dataset

BOX PLOT

```python
plt.subplots(figsize=(16,12))
sns.boxplot(data=df,x="total",y="not_distracted");
```

JOINT PLOT

```python
sns.jointplot(data=df,x="speeding",y="total");
```

INFERENCE: It can be observed that even at the average speeding the total crashes were more than high speeding

COUNT PLOT

```python
plt.subplots(figsize=(16,10))
sns.countplot(data=df,y="total",orient='h');
```

The total are more at 23.9, 21.4, 14.1,12.8,13.6
